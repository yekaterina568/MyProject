```python
import pygame
import sys
import random

pygame.init()

# Параметры экрана и клетки
width, height = 500, 500
cell_size = 10

# Цвета
black = (0, 0, 0)
green = (0, 255, 0)
red = (255, 0, 0)
white = (255, 255, 255)

# Окно
screen = pygame.display.set_mode((width, height))
pygame.display.set_caption('Snake Game')

# Начальные параметры змейки
snake_pos = [100, 100]
snake_body = [[100, 100], [90, 100], [80, 100]]
direction = 'RIGHT'
change_to = direction

# Начальная еда 
food_pos = [random.randrange(1, width // cell_size) * cell_size,
            random.randrange(1, height // cell_size) * cell_size]
food_spawn = True

# Счёт и уровень
score = 0
level = 1
speed = 10

# Шрифт
font = pygame.font.SysFont('Verdana', 18)

# Таймер
clock = pygame.time.Clock()

# Проверка столкновения с телом или стеной
def is_collision(pos):
    if pos in snake_body[1:]:  # Столкновение с телом
        return True
    if pos[0] < 0 or pos[0] >= width or pos[1] < 0 or pos[1] >= height:  # Столкновение со стеной
        return True
    return False

# Генерация новой еды 
def generate_food():
    while True:
        new_pos = [random.randrange(1, width // cell_size) * cell_size,
                   random.randrange(1, height // cell_size) * cell_size]
        if new_pos not in snake_body:
            return new_pos

# Главный цикл
running = True
while running:
    # Управление направлением
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and direction != 'DOWN':
                change_to = 'UP'
            elif event.key == pygame.K_DOWN and direction != 'UP':
                change_to = 'DOWN'
            elif event.key == pygame.K_LEFT and direction != 'RIGHT':
                change_to = 'LEFT'
            elif event.key == pygame.K_RIGHT and direction != 'LEFT':
                change_to = 'RIGHT'

    direction = change_to

    # Обновление позиции головы
    if direction == 'UP':
        snake_pos[1] -= cell_size
    elif direction == 'DOWN':
        snake_pos[1] += cell_size
    elif direction == 'LEFT':
        snake_pos[0] -= cell_size
    elif direction == 'RIGHT':
        snake_pos[0] += cell_size

    # Проверка на столкновение
    if is_collision(snake_pos):
        running = False 

    # Перемещение змейки
    snake_body.insert(0, list(snake_pos))

    # Проверка еды
    if snake_pos == food_pos:
        score += 1
        if score % 4 == 0:
            level += 1
            speed += 2  # Увеличение скорости
        food_pos = generate_food()
    else:
        snake_body.pop()

    # Отрисовка
    screen.fill(black)
    for block in snake_body:
        pygame.draw.rect(screen, green, pygame.Rect(block[0], block[1], cell_size, cell_size))

    pygame.draw.rect(screen, red, pygame.Rect(food_pos[0], food_pos[1], cell_size, cell_size))

    # Отображение счёта и уровня
    score_text = font.render(f'Score: {score}', True, white)
    level_text = font.render(f'Level: {level}', True, white)
    screen.blit(score_text, (10, 10))
    screen.blit(level_text, (400, 10))

    pygame.display.flip()
    clock.tick(speed)

# Завершение
pygame.quit()
```
