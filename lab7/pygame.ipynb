```python
import pygame
import sys
import time

pygame.init()

WIDTH, HEIGHT = 800, 800
screen = pygame.display.set_mode((WIDTH, HEIGHT))
clock = pygame.time.Clock()

background = pygame.image.load("mickey_face.png").convert()
right_hand = pygame.image.load("right_hand.png").convert_alpha()
left_hand = pygame.image.load("left_hand.png").convert_alpha()

background = pygame.transform.scale(background, (WIDTH, HEIGHT))
right_hand = pygame.transform.scale(right_hand, (400, 200))
left_hand = pygame.transform.scale(left_hand, (400, 200))

right_hand_pos = (WIDTH // 2, HEIGHT // 2)
left_hand_pos = (WIDTH // 2, HEIGHT // 2)

running = True
while running:
    screen.blit(background, (0, 0))

    current_time = time.localtime()
    minute_angle = -((current_time.tm_min % 60) * 6)
    second_angle = -((current_time.tm_sec % 60) * 6)

    rotated_right = pygame.transform.rotate(right_hand, minute_angle)
    rotated_left = pygame.transform.rotate(left_hand, second_angle)

    right_rect = rotated_right.get_rect(center=right_hand_pos)
    left_rect = rotated_left.get_rect(center=left_hand_pos)

    screen.blit(rotated_right, right_rect.topleft)
    screen.blit(rotated_left, left_rect.topleft)

    pygame.display.flip()
    clock.tick(30)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

pygame.quit()
sys.exit()

```

```python
import pygame
import os

pygame.init()
pygame.mixer.init()

WIDTH, HEIGHT = 400, 300
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Music Player")

MUSIC_FOLDER = "music"
tracks = [f for f in os.listdir(MUSIC_FOLDER) if f.endswith(".mp3")]
current_track = 0

font = pygame.font.Font(None, 36)

def draw_button(text, x, y, w, h, color, text_color):
    pygame.draw.rect(screen, color, (x, y, w, h))
    label = font.render(text, True, text_color)
    screen.blit(label, (x + 10, y + 10))

def play_music():
    pygame.mixer.music.load(os.path.join(MUSIC_FOLDER, tracks[current_track]))
    pygame.mixer.music.play()

def stop_music():
    pygame.mixer.music.stop()

def next_track():
    global current_track
    current_track = (current_track + 1) % len(tracks)
    play_music()

def prev_track():
    global current_track
    current_track = (current_track - 1) % len(tracks)
    play_music()

if tracks:
    play_music()

running = True
while running:
    screen.fill((30, 30, 30))
    
    draw_button("Play", 50, 200, 80, 50, (0, 200, 0), (255, 255, 255))
    draw_button("Stop", 150, 200, 80, 50, (200, 0, 0), (255, 255, 255))
    draw_button("Next", 250, 200, 80, 50, (0, 0, 200), (255, 255, 255))
    draw_button("Prev", 50, 260, 80, 50, (200, 200, 0), (0, 0, 0))
    
    pygame.display.flip()
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                play_music()
            elif event.key == pygame.K_s:
                stop_music()
            elif event.key == pygame.K_n:
                next_track()
            elif event.key == pygame.K_p:
                prev_track()
        elif event.type == pygame.MOUSEBUTTONDOWN:
            x, y = event.pos
            if 50 <= x <= 130 and 200 <= y <= 250:
                play_music()
            elif 150 <= x <= 230 and 200 <= y <= 250:
                stop_music()
            elif 250 <= x <= 330 and 200 <= y <= 250:
                next_track()
            elif 50 <= x <= 130 and 260 <= y <= 310:
                prev_track()

pygame.quit()
```

```python
import pygame

pygame.init()

WIDTH, HEIGHT = 500, 500
BALL_RADIUS = 25
STEP = 20

screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Moving Ball")

ball_x, ball_y = WIDTH // 2, HEIGHT // 2
running = True

while running:
    screen.fill((255, 255, 255))
    pygame.draw.circle(screen, (255, 0, 0), (ball_x, ball_y), BALL_RADIUS)
    pygame.display.flip()
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_UP and ball_y - BALL_RADIUS - STEP >= 0:
                ball_y -= STEP
            elif event.key == pygame.K_DOWN and ball_y + BALL_RADIUS + STEP <= HEIGHT:
                ball_y += STEP
            elif event.key == pygame.K_LEFT and ball_x - BALL_RADIUS - STEP >= 0:
                ball_x -= STEP
            elif event.key == pygame.K_RIGHT and ball_x + BALL_RADIUS + STEP <= WIDTH:
                ball_x += STEP

pygame.quit()
```
